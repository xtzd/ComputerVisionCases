```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img1 = cv2.imread('images/1-1.PNG')
img2 = cv2.imread('images/1-2.PNG')
```

```python
plt.subplot(121)
plt.imshow(cv2.cvtColor(img1,cv2.COLOR_BGR2RGB))
plt.subplot(122)
plt.imshow(cv2.cvtColor(img2,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
img12 = cv2.imread('1-output.JPG')
plt.imshow(cv2.cvtColor(img12,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
sift = cv2.xfeatures2d.SIFT_create()
```

```python
pt1, des1 = sift.detectAndCompute(cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY),None)
pt2, des2 = sift.detectAndCompute(cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY),None)
```
